### Importing & Cleaning Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('puchotrain.csv')
train.head(1)
```

```python
test=pd.read_csv('puchotest.csv')
test.head(1)
```

```python
train.columns
```

```python
X_train=train.drop('TARGET',axis=1)
y_train=train['TARGET']
```

```python
test.columns
```

```python
for i in train.columns:
    print(train[i].value_counts())
```

```python
for i in test.columns:
    print(test[i].value_counts())
```

```python
train = train.replace(-999999,2)
```

### Feature Engineering


```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
clf = ExtraTreesClassifier(random_state=0)
clf = clf.fit(X_train, y_train)
clf.feature_importances_  

model = SelectFromModel(clf, prefit=True)
```

```python
X_train_opt = model.transform(X_train)
feature_idx_train = model.get_support()
feature_name_train = X_train.columns[feature_idx_train]
print(feature_name_train)
```
